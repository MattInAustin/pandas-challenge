```python
#Import stuff
import os
import csv
import pandas as pd
import numpy as np
```

```python
#load the resource files
schools_sheet = os.path.join("Resources", "schools_complete.csv")
students_sheet = os.path.join("Resources", "students_complete.csv")

#Read school file
schools_df = pd.read_csv(schools_sheet)
students_df = pd.read_csv(students_sheet)


school_combined = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
```

```python
#create a table with "Total schools, total students, total budget, average math score, average reading score, % passing math, %passing reading, and overall passing rate (average of the last two)"

#Count # of schools
count_schools = len(school_combined['school_name'].unique())

#Count students
count_students = school_combined['student_name'].count()

#Total budget
total_budget = sum(school_combined['budget'].unique())

#Average math score
average_math = school_combined['math_score'].mean()

#Average reading score
average_reading = school_combined['reading_score'].mean()

#Passing rate
passing_rate = (average_math + average_reading) / 2

#Passing Math
passing_math = (school_combined[school_combined['math_score']>=70]['student_name'].count()/count_students)*100

#Passing Reading
passing_reading = (school_combined[school_combined['reading_score']>=70]['student_name'].count()/count_students)*100

#Put the above variables into dataframe

summary = {
    'Total Schools':count_schools,
    'Total Students':count_students,
    'Total Budget':total_budget,
    'Average Math Score':average_math,
    'Average Reading Score':average_reading,
    '% Passing Math':passing_math,
    '% Passing reading':passing_reading,
    '% Passing Total':[passing_rate],
}

display_summary_dataframe = pd.DataFrame(summary)
display_summary_dataframe

#count_school_names = schools_sheet['school'].unique()
```

```python
#Group by school name
groupby_school = school_combined.groupby(['school_name'])

#Total student for school
student_total = groupby_school.size()

#School type
type_school = groupby_school['type'].first()

#Calculate budget
budget_total = groupby_school['budget'].first()

#Calculate the budget per student
budget_per_student = budget_total/student_total

#Average math score
math_average = groupby_school['math_score'].mean()

#Calculate average reading score
reading_average = groupby_school['reading_score'].mean()

#Calculate the percentage of passing math score
passing_math = school_combined[school_combined['math_score']>=70].groupby(['school_name']).size()
percent_math = (passing_math/student_total) * 100

#Calculate percentage of passing reading score
passing_reading = school_combined[school_combined['reading_score']>=70].groupby(['school_name']).size()
percent_reading = (passing_reading/student_total) * 100

#Overall passing score
overall_passing = (percent_math + percent_reading) / 2

#Dataframe to put above values into
school_summary_df = {
    'School Type': type_school,
    'Total Students':student_total,
    'Total School Budget':budget_total,
    'Per Student Budget': budget_per_student,
    'Average Math Score': math_average,
    'Average Reading Score': reading_average,
    '% Passing Math': percent_math,
    '% Passing Reading':percent_reading,
    '% Passing Overall': overall_passing
}

display_summary_school = pd.DataFrame(school_summary_df)
display_summary_school

```

```python
# Bins Spending
bins_spending = [0, 585, 600, 630, 680]
names_group = ["585", "600", "630", "680"]

#Make a new dataframe
school_spendingdf = pd.DataFrame(school_summary_df)

#cut each school up by budget and student
school_spendingdf['Spending Ranges (Per Student)'] = pd.cut(school_spendingdf['Per Student Budget'], bins_spending, labels = names_group)

#Group by Spending Rangers per student
spending_per_group = school_spendingdf.groupby('Spending Ranges (Per Student)')

#Calculate the averages
stu_avg_math = spending_per_group['Average Math Score'].mean()
stu_avg_reading = spending_per_group['Average Reading Score'].mean()
stu_pct_math = spending_per_group['% Passing Math'].mean()
stu_pct_reading = spending_per_group['% Passing Reading'].mean()
stu_pct_over = spending_per_group['% Passing Overall'].mean()


spending_per_student_dict = {
    'Average Math Score': stu_avg_math,
    'Average Reading Score': stu_avg_reading,
    '% Passing Math': stu_pct_math,
    '% Passing Reading': stu_pct_reading,
    '% Passing Overall': stu_pct_over,
}


spending_per_studentdf = pd.DataFrame(spending_per_student_dict)

spending_per_studentdf
```
